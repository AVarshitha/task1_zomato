```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

# Set styling for neat visualizations
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

print("Libraries successfully imported!")
```

```python
# Load the dataset
df = pd.read_csv('zomato.csv')

# Display basic information to understand our columns
print("--- Dataset Info ---")
print(df.info())
print("\n--- Missing Values Count ---")
print(df.isnull().sum())
```

```python
# 1. Handling Missing Values
# Updated to use 'approx_cost(for two people)' to fix the KeyError
df.dropna(subset=['rate', 'approx_cost(for two people)', 'cuisines'], inplace=True)

# 2. Robust Cleaning for the 'rate' (Rating) column
def clean_rating(value):
    if pd.isna(value):
        return np.nan

    # Clean up the string (removes extra spaces or hidden characters like \n)
    value = str(value).split('/')[0].strip()

    if value.upper() in ['NEW', '-', '']:
        return np.nan

    # Try-Except block: Handles random text strings like '\nTop floor' safely
    try:
        return float(value)
    except ValueError:
        return np.nan  # If it's text that can't be a number, turn it into NaN safely

df['rate'] = df['rate'].apply(clean_rating)
df.dropna(subset=['rate'], inplace=True) # Drop rows that failed the numeric conversion

# 3. Cleaning the 'approx_cost(for two people)' column
# Fixed the typo here to match your exact dataset column name
df['approx_cost(for two persons)'] = df['approx_cost(for two people)'].astype(str).str.replace(',', '')
df['approx_cost(for two persons)'] = pd.to_numeric(df['approx_cost(for two persons)'], errors='coerce')
df.dropna(subset=['approx_cost(for two persons)'], inplace=True)

# 4. Standardizing text data
df['name'] = df['name'].str.strip()
df['location'] = df['location'].str.strip()

print("Data Cleaning complete. Shape of cleaned dataset:", df.shape)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set up a side-by-side layout for both plots
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Plot 1: Price vs. Rating Scatter Plot
sns.scatterplot(ax=axes[0], data=df, x='approx_cost(for two persons)', y='rate')
axes[0].set_title('Price vs. Rating')

# Plot 2: Top 10 Restaurant Locations
top_locations = df['location'].value_counts().head(10)
sns.barplot(ax=axes[1], x=top_locations.values, y=top_locations.index)
axes[1].set_title('Top 10 Restaurant Locations')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Select only numeric columns
numeric_df = df.select_dtypes(include=[np.number])

# Generate heatmap with the exact heading name
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Combine all cuisines into one text string
cuisine_text = " ".join(df['cuisines'].dropna().astype(str))

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(cuisine_text)

# Display the plot with the heading name
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Cuisine Word Cloud')
plt.show()
```
